```python

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
import rioxarray as rxr
from rasterio.plot import plotting_extent
import earthpy as et

# Mikes Fucntions
import index_functions as index
from get_silo import monthly_rain

sns.set(font_scale=1.5, style="white")
```

```python
sample_band = rxr.open_rasterio('landsat/sowing/LC08_L1TP_111082_20190525_20190605_01_T1_B02.TIF')

lga = gpd.read_file('LGA/LGA.shp')

sample_shape = lga[lga['LGA_NAME20'] == 'Wagin (S)']
```

```python
rain_test = monthly_rain(2019, 
                         from_month = 1, 
                         x_months = 5, 
                         proj = sample_band,
                         bound = sample_shape)
 
rain_test
 
```

```python
f, ax = plt.subplots(figsize = (18, 12))

rain_test.plot.imshow(cmap = "viridis",
                      ax = ax)

ax.set_title('SILO Raster Example \n Wagin Cumulative Rainfalll \n Jan-May 2019', fontdict = {'fontsize': 28, 'fontweight': 'medium'})
ax.axis('off')
plt.show()

```
